## 3.1 Label encoding across multiple columns in scikit-learn
As you can see the above, the dataframe has many columns with none binary categories(this can not be converted into 1/0 (binary) values).it means, If I choose the approach that creates a LabelEncoder object for each colum, there are so tedious works to do for me. 
 SO, I decide to avoid creating a LabelEncoder object for each column. My choice is to create a multi-column LabelEncoder() instead.

```python

df = df.fillna('0', axis=0)
```

```python
# this solution for applying LabelEncoder() across multiple columns was suggested in the following thread
# http://stackoverflow.com/questions/24458645/label-encoding-across-multiple-columns-in-scikit-learn
class MultiColumnLabelEncoder:
    def __init__(self,columns = None):
        self.columns = columns 

    def fit(self,X,y=None):
        return self

    def transform(self,X):
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = LabelEncoder().fit_transform(output[col])
        else:
            for colname,col in output.iteritems():
                output[colname] = LabelEncoder().fit_transform(col)
        return output

    def fit_transform(self,X,y=None):
        return self.fit(X,y).transform(X)
```

```python
def processor(data):
    data = MultiColumnLabelEncoder(columns = ['people_id', 'activity_id', 'date', 'activity_category', 'char_1',\
                                              'char_2', 'char_3', 'char_4', 'char_5', 'char_6', 'char_7',\
                                              'char_8', 'char_9', 'char_10']).fit_transform(df)
    

    
    return data
```

# 5.1 GridSearch continue

So far, I ran the descision tree using defaults for the parameters. These could be the best choices for the parameters or not. In case of large number of parameters to test, one way to improve my scores is with a GridSearch.
This technique allows us to search through all combinations of a set of parameters for a model in order to determine which combination is most effective. The parameters of the model are optimized by a cross-validated grid search over a parameter grid.

In scikit-learn, this technique is provided in the GridSearchCV class.

When constructing this class you must provide a dictionary of hyperparameters to evaluate in the param_grid argument. This is a map of the model parameter name and an array of values to try.

By default, accuracy is the score that is optimized, but other scores can be specified in the score argument of the GridSearchCV constructor.

By default the grid search will only use one thread. By setting the n_jobs argument in the GridSearchCV constructor to -1, the process will use all cores on the machine.

The GridSearchCV process when then construct and evaluate one model for each combination of parameters. Cross validation is used to evaluate each individual model and the default of 3-fold cross validation is used, although this can be overridden by specifying the cv argument to the GridSearchCV constructor.

```python
from sklearn import grid_search
# creat a dict with all of the parameters
param_grid = {'min_samples_split' : [1, 5, 10, 15, 20, 25, 30],
              'max_depth' : [4, 5, 6, 7, 8]}

# create GridSearch object with param_grid
gridCV_object = grid_search.GridSearchCV(estimator = DecisionTreeClassifier(random_state = 53),
                                         param_grid = param_grid,
                                         cv = 5)

#fit to the data
gridCV_object.fit(X_train, y_train)

print "Best parameters from parameter grid:"
pprint(gridCV_object.best_params_)

#  the best estimator
clf_2 = gridCV_object.best_estimator_

# best param for the best estimator.
print "\nComplete set of parameters for best estimator:"
pprint(clf_2.get_params())

# check scores
print "\nAccuracy score for default, decision tree:"
print clf.score(X_test, y_test)

print "\nAccuracy score for best estimator from gridSearch"
print clf_2.score(X_test, y_test)
```

# 5.2 Pipelines

A pipeline is a complete classifier object. it can be used with .fit() and .predict()

Pipelines allow users apply a list of transforms together with a final estimator and is very useful in case of varying the parameters of transform as well as the params of the classifier.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Imputer
from sklearn.decomposition import PCA

pipeline = Pipeline([('imp', Imputer()),
                     ('std', StandardScaler()),
                     ('pca', PCA()),
                     ('clf', DecisionTreeClassifier(random_state = 53)),
                    ])

pipeline = pipeline.fit(X_train, y_train)

print "Pipeline Accuracy Score"
print pipeline.score(X_test, y_test)

print "\nPipeline params:"
pprint(pipeline.get_params())
```

# 5.3 Pipeline and GridSearchCV together

```python

# creat a dict with all of the parameters
param_grid = {'imp__strategy': ['median'],
              'clf__min_samples_split': [1, 5, 10, 15, 20, 25, 30],
              'clf__max_depth': [4, 5, 6, 7, 8],
              'pca__n_components':[2, 3, 4, 5, 6, 7]}
# create GridSearch object with param_grid
gridCV_object = grid_search.GridSearchCV(estimator = pipeline,
                                         param_grid = param_grid,
                                         cv = 8)

#fit to the data
gridCV_object.fit(X_train, y_train)

print "Best parameters from parameter grid:"
pprint(gridCV_object.best_params_)

#  the best estimator
pipeline_clf = gridCV_object.best_estimator_



# check scores
print "\nBest estimator accuracy score:"
print pipeline_clf.score(X_test, y_test)

```
